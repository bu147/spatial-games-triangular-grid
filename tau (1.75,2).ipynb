```python
from Assemblies.game import GameField
import numpy as np
from numpy import random
import time

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
b=7/4+0.01
percentage1=21
percentage2=50
size=72
burnin=10000
evol=40000
```

```python
#возвращает среднее время смены стратегии
def tau(size, b, percentage=21, seed=1, burnin=1000, evol=50):
    random.seed(seed)
    instance = np.array(random.rand(size, size) > 0.01 * percentage, dtype=int)
    
    game = GameField(size, b)
    game.field = np.copy(instance)
    
    game.evolve(burnin)
    
    #снимки шагов системы
    snapshots=np.zeros((size,size,evol), dtype=int)
    snapshots[:,:,0]=game.field
    
    for step in range(evol):
        game.evolve(1)
        snapshots[:,:,step]=game.field
    #-----------
    
    #сколько раз каждый игрок менял стратегию
    switches=np.zeros_like(snapshots[:,:,0], dtype=int)
    
    for i in range(snapshots.shape[0]):
        for j in range(snapshots.shape[1]):
            strats=snapshots[i,j,:] #все стратегии одного игрока
            idx,=(strats[1:]-strats[:-1]).nonzero() #шаги игры, на которых игрок менял стратегию
            switches[i,j]=idx.size+1 #количество смен стратегий
    #-----------
    
    #количество игроков, менявших стратегию
    swingers=switches>1
    num_swingers=np.count_nonzero(swingers)
    swing_perc=(num_swingers/size**2) * 100
    #-----------
    
    #число шагов системы / количество смен стратегий игрока = время придерживания определённой стратегии для одного игрока
    #выбираем только игроков, менявших стратегию и берём среднее
    tau_swingers=np.mean(evol/switches[swingers])
    #-----------
    
    print('seed:', seed)
    print('среднее время смены стратегии: ', tau_swingers)
    print('количество игроков, менявших стратегию: ', num_swingers)
    print('процент игроков, менявших стратегию: %.2f' % swing_perc)
    print('-----------')
    return tau_swingers
```

```python
print('размер поля: ',size,'\nпараметр выигрыша: ',b,'\nшаги отжига: ',burnin, '\nшаги эволюции: ',evol,'\n-----------')
seeds=np.array([1,56,2341,98,501,1004,20450,7,403,1340018])
ss=seeds.size

print('===========начальный процент кооператоров: ',percentage1,'===========')
tests1=np.zeros(ss)
for i in range(ss):
    tests1[i]=tau(size, b, percentage1, seeds[i], burnin, evol)
    

print('===========начальный процент кооператоров: ',percentage2,'===========')
tests2=np.zeros(ss)
for i in range(ss):
    tests2[i]=tau(size, b, percentage2, seeds[i], burnin, evol)
```
